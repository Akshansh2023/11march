Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.


 # Answer 
 Both t-tests and z-tests are hypothesis tests used to make inferences about population parameters based on sample statistics. The main difference between them lies in the assumptions made about the population variance.

A z-test assumes that the population variance is known, whereas a t-test assumes that the population variance is unknown and estimates it from the sample variance. In general, z-tests are used when the sample size is large (usually greater than 30) and the population variance is known, whereas t-tests are used when the sample size is small (usually less than 30) or the population variance is unknown.

Example scenario for a z-test: Suppose you want to test whether the average weight of all apples produced by a farm is 200 grams. You have a large sample of 200 apples and know the population variance to be 25 grams. In this case, you could use a z-test to determine whether the average weight of the sample is significantly different from 200 grams.

Example scenario for a t-test: Suppose you want to test whether a new drug reduces blood pressure. You have a sample of 25 patients and do not know the population variance. In this case, you could use a t-test to determine whether the average reduction in blood pressure in the sample is significantly different from zero.

Q2: Differentiate between one-tailed and two-tailed tests.

#  Answer 

In hypothesis testing, a one-tailed test and a two-tailed test refer to the directionality of the alternative hypothesis.

A one-tailed test is a hypothesis test where the alternative hypothesis states the direction of the effect, either an increase or a decrease in a particular outcome variable. In other words, it only considers one direction of change. The null hypothesis represents the opposite direction. For example, we might conduct a one-tailed test to determine if a new drug increases the average response time of participants.

On the other hand, a two-tailed test is a hypothesis test where the alternative hypothesis does not specify the direction of the effect. It considers two directions of change, either an increase or a decrease in a particular outcome variable. The null hypothesis in this case is that there is no effect. For example, we might conduct a two-tailed test to determine if a new drug affects the average response time of participants.

The choice of one-tailed or two-tailed test depends on the research question and the nature of the hypothesis. One-tailed tests are used when we have a specific directional hypothesis and want to test if the effect is in that specific direction. Two-tailed tests are used when the hypothesis does not specify the direction of the effect or when there is no prior knowledge about the direction of the effect.

In summary, the key difference between one-tailed and two-tailed tests is the directionality of the alternative hypothesis. One-tailed tests consider only one direction of change, while two-tailed tests consider two directions of change.

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

# Answer 

In hypothesis testing, Type 1 and Type 2 errors are two possible errors that can occur when testing a null hypothesis.

Type 1 error occurs when a true null hypothesis is rejected. This is also known as a false positive. The probability of committing a Type 1 error is denoted by the Greek letter alpha (α), and is typically set at 0.05 or 0.01.

Example scenario for Type 1 error: A pharmaceutical company is testing a new drug to reduce cholesterol levels. They set a null hypothesis that the drug has no effect on cholesterol levels. They conduct a clinical trial and conclude that the drug is effective at reducing cholesterol levels when, in reality, it has no effect.

Type 2 error occurs when a false null hypothesis is not rejected. This is also known as a false negative. The probability of committing a Type 2 error is denoted by the Greek letter beta (β).

Example scenario for Type 2 error: A medical researcher is conducting a study on the effectiveness of a new cancer treatment. They set a null hypothesis that the treatment is not effective. In reality, the treatment is effective, but the researcher concludes that it is not based on their study.

Both Type 1 and Type 2 errors are important to consider in hypothesis testing as they can have significant consequences. Choosing an appropriate sample size, using the correct statistical test, and setting appropriate levels of alpha and beta can help to reduce the risk of making these errors.


Q4: Explain Bayes's theorem with an example.
# Answer
Bayes's theorem is a statistical formula that allows us to update our beliefs or knowledge about the probability of an event occurring based on new information or evidence. It relates the conditional probability of an event A given an event B to the conditional probability of event B given event A.

Bayes's theorem can be written as follows:

P(A|B) = P(B|A) * P(A) / P(B)

where:

P(A|B) is the conditional probability of event A given event B
P(B|A) is the conditional probability of event B given event A
P(A) is the prior probability of event A
P(B) is the prior probability of event B
Here's an example to illustrate Bayes's theorem:

Suppose that you want to test for a rare genetic disorder in a population. The prevalence of the disorder is 1 in 10,000, meaning that the prior probability of someone having the disorder is 0.0001. You conduct a diagnostic test that has a sensitivity of 95% and a specificity of 99%. This means that the test correctly identifies 95% of people who have the disorder (true positive rate) and correctly identifies 99% of people who do not have the disorder (true negative rate).

Now suppose that you administer the test to a randomly selected person and the test comes back positive. What is the probability that this person actually has the genetic disorder?

Using Bayes's theorem, we can update our prior probability of the person having the disorder based on the new evidence (positive test result) as follows:

P(A|B) = P(B|A) * P(A) / P(B)

where:

A is the event that the person has the genetic disorder
B is the event that the test is positive
We can calculate P(B|A) as the sensitivity of the test, which is 0.95, and P(A) as the prior probability of having the disorder, which is 0.0001. P(B) can be calculated using the law of total probability:

P(B) = P(B|A) * P(A) + P(B|~A) * P(~A)

where ~A is the complement of event A, meaning the person does not have the disorder. We can calculate P(B|~A) as the false positive rate of the test, which is 1 - specificity, or 0.01, and P(~A) as 1 - P(A), which is 0.9999.

Plugging in the values, we get:

P(A|B) = 0.95 * 0.0001 / (0.95 * 0.0001 + 0.01 * 0.9999) ≈ 0.0094

This means that the probability of the person actually having the genetic disorder given a positive test result is only about 0.94%. Even though the test has a high sensitivity and specificity, the rarity of the disorder means that a positive test result is still more likely to be a false positive than a true positive.







Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

# ANSWER
A confidence interval is a range of values that is likely to contain the true value of a population parameter with a certain level of confidence. It is a measure of the precision or uncertainty in the estimate of the population parameter.

For example, suppose we want to estimate the average weight of all female college students in the United States. We could take a random sample of 100  female college students and calc
ulate the sample mean weight. However, we know that the sample mean weight may not be exactly equal to the true population mean weight, due to random sampling error. A confidence interval provides us with a range of values within which we can be reasonably confident that the true population mean weight lies.

To calculate a confidence interval, we need to know the sample size, the sample statistic (e.g. mean, proportion), the standard deviation (or standard error), and the level of confidence. The most common levels of confidence are 95% and 99%.

The formula for calculating a confidence interval for a population mean is:

Confidence interval = sample mean ± (z-score * standard error)

where the z-score is obtained from the standard normal distribution table and corresponds to the level of confidence. For example, for a 95% confidence level, the z-score is 1.96. The standard error is the estimated standard deviation of the sampling distribution of the sample mean, which is calculated as the sample standard deviation divided by the square root of the sample size.

Let's use the above example to illustrate how to calculate a confidence interval. Suppose we take a random sample of 100 female college students and measure their weight. The sample mean weight is 130 pounds, with a sample standard deviation of 20 pounds. We want to calculate a 95% confidence interval for the population mean weight.

Using the formula above, we can calculate the standard error as:

standard error = sample standard deviation / sqrt(sample size)
= 20 / sqrt(100)
= 2

The z-score for a 95% confidence level is 1.96, so the confidence interval is:

Confidence interval = sample mean ± (z-score * standard error)
= 130 ± (1.96 * 2)
= (126.08, 133.92)

This means that we are 95% confident that the true population mean weight lies within the range of 126.08 to 133.92 pounds.

t = (x̄ - μ) / (s / √n)

where:

x̄ is the sample mean (6 pounds)
μ is the population mean (0 if the null hypothesis is true)
s is the sample standard deviation (2.5 pounds)
n is the sample size (50)
We will use a significance level (α) of 0.05, which corresponds to a 95% confidence level. We need to find the critical value for a one-tailed t-test with 49 degrees of freedom (df = n - 1) using a t-table or a calculator. The critical value for a 95% confidence level and 49 degrees of freedom is approximately 1.677.

Calculating the t-statistic:

t = (6 - 0) / (2.5 / √50)
t = 16.97

The t-statistic is much larger than the critical value, so we reject the null hypothesis and conclude that the weight loss drug is significantly effective at a 95% confidence level.

Therefore, we can conclude that the weight loss drug is significantly effective at a 95% confidence level.
